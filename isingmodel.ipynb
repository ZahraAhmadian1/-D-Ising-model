```python
import numpy as np
import matplotlib.pyplot as plt

#number of total spins
N=200
J=1
KT=1
#number of steps
step=1000
t=[]
#initial conditions
spins=np.random.choice([-1,1],N)


def energy_cal(spins):
    calene=0.0
    for j in range(N):
        if j == 0:
            ene=-J*(spins[-1]*spins[j]+spins[j+1]*spins[j])
        elif j== N-1:
            ene=-J*(spins[j-1]*spins[j]+spins[0]*spins[j])
        else:
            ene=-J*(spins[j-1]*spins[j]+spins[j+1]*spins[j])
        calene=calene+ene
               
    return calene
        
    
magnes=[]
energy=[]
for i in range(step):
    
    ran=np.random.randint(N)
    s=spins[ran]
    if ran == 0:
        ene=2*J*(spins[-1]*spins[ran]+spins[ran+1]*spins[ran])
    elif ran== N-1:
        ene=2*J*(spins[ran-1]*spins[ran]+spins[0]*spins[ran])
    else:
        ene=2*J*(spins[ran-1]*spins[ran]+spins[ran+1]*spins[ran])

    if ene < 0 or np.random.rand() < np.exp(-ene/KT):
        spins[ran]=-s

    if i%10 == 0:
        magnes.append(np.sum(spins))
        energy.append(energy_cal(spins))

t=np.arange(len(magnes))

plt.subplot(1,2,1)
plt.plot(t,energy)
plt.xlabel('time')
plt.ylabel('energy')

plt.subplot(1,2,2)
plt.plot(t,magnes)
plt.xlabel('time')
plt.ylabel('mangnetization')
plt.tight_layout()




        
    
    
```
